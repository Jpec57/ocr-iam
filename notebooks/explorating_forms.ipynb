```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
forms_meta_filename = "../data/ascii/forms.txt"

def get_first_dataset_line_index(filename):
    with open(filename,'r') as f:
        i = 1
        for line in f:
            if not line.startswith("#"): 
                return i
            i += 1
        return -1

first_line_index = get_first_dataset_line_index(forms_meta_filename)

columns = ['form_id', 'writer_id', 'number_of_sentences', 'word_seg', 'total_lines', 'correct_lines', 'total_words', 'correct_words']
df = pd.read_csv(forms_meta_filename, header=first_line_index-1, delimiter=" ", names=columns)
df.head(50)
```
